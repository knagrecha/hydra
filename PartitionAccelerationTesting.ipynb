```python
import torch
import torch.nn as nn
```

```python
tensor = torch.ones(5000, 100, 100)
```

```python
tensor = tensor.to("cuda:0")
```

```python
from hydra.utilities import get_free_space 
```

```python
get_free_space(0)
```

```python
from timeit import default_timer as timer
```

```python
partitions = 1000
```

```python
start = timer()
partition_list = []
multiplier = int(tensor.shape[0]/partitions)

for x in range(1, partitions):
    val = tensor[multiplier*(x-1):multiplier*x, :, :].clone()
    partition_list.append(val)

print("Time taken: {:.2f}".format(timer() - start))
```

```python
print(partition_list[0].shape)
```

```python
start = timer()
with torch.no_grad():
    torch.cat(partition_list, dim=0)

print("Time taken: {:.2f}".format(timer() - start))
```
